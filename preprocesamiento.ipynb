```python
import pandas as pd
goles=pd.DataFrame([3,2,3,1,0,0,2,3,1,8])
print("Media: ",goles.mean(),"\nMediana: ", goles.median(),"\nModa: ", goles.mode())
print("Desviación típica:", goles.std())
print("Varianza:", goles.var())
print("Max", goles.max())
print("Min:", goles.min())
print("Varianza:", goles.var())
print("Quantil 1: ", goles.quantile(0.25)) # Cálculo de cuantiles, 25%, 50%, 75%
print("Quantil 3:", goles.quantile(0.75)) # Quantil 3
```

```python
import numpy as np
 
# Contexto: levantamiento de pesas
pesos = [5, 15, 25, 35, 40, 55, 80, 90, 170]
print("Rango:", np.max(pesos) - np.min(pesos))
print("Varianza poblacional:", np.var(pesos))
#Varianza muestral. Se resta a los datos 1, por eso se pone ddof=1
#Si no se pone nada se asume 0, sería poblacional
print("Varianza muestral:", np.var(pesos, ddof=1))
print("Desviación estándar poblacional:", np.std(pesos))
print("Desviación estándar muestral:", np.std(pesos, ddof=1))

#Cálculo del rango intercuartílico
Q1 = np.percentile(pesos, 25)
Q3 = np.percentile(pesos, 75)
print("Rango intercuartílico:", Q3 - Q1)
```

```python
datos = np.random.normal(size=1000000)
plt.hist(datos, color="pink", ec="black", alpha=0.5) #ejemplo típico de las notas de exámenes, o de las alturas de las personas.
plt.title("Distribución Normal")
plt.show()
```

```python
datos = np.random.uniform(size=1000000) # Cada uno de los valores tiene una frecuencia uniforme, lotería nacional, cada valor tiene la misma probabilidad de aparecer
plt.hist(datos, color="skyblue", ec="black", alpha=0.5)
plt.title("Distribución Uniforme")
plt.show()

datos = np.random.exponential(size=1000000)
plt.hist(datos, color="green", ec="black", alpha=0.5) #tiempo en fallar un ordenador, tiempo que las personas pasan en un sitio web.
plt.title("Distribución Exponencial")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

# altura promedio
mu = 170 

# desviación estándar
sigma = 8 

# generación de 100 mil alturas aleatorias, siguiendo una distribución normal
n = 100000
alturas = np.random.normal(mu, sigma, n)

# Creación de histograma, 100 beans
plt.figure(figsize=(9,5))
conteo, x, barras = plt.hist(alturas, 100, density=True, color="pink")
```

```python
conteo, x, barras = plt.hist(alturas, 100, density=True, color="pink")
print(conteo,"\n")
print(x,"\n")
print(barras,"\n")
```

```python

# Calculamos media, mediana y moda. La media y mediana será iguales a 170 +-
media=alturas.mean()
mediana=np.median(alturas)
# Numpy no tiene moda
from scipy import stats as st
moda=st.mode(alturas)
print("Media: ", media, "Mediana: ", mediana,"Moda: ", moda)

# Utilizando x, con los bordes de los bins
# Dibujamos la campana utilizando la función de densidad de probabilidad
plt.plot(x,  1 / (sigma*np.sqrt(2*np.pi)) * np.exp(-0.5*((x-mu)/sigma)**2),
         linewidth=1, color="purple")
```
