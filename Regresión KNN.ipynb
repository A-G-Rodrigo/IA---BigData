```python
# Regresión lineal
# ---import pandas as pd
# ---from google.colab import files
# ---files.upload()
datos = pd.read_csv("datasets/datasets/ingreso.csv")
datos
```

```python
import matplotlib.pyplot as plt
plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="pink")
plt.show()
```

```python
# Creación del modelo 
# Regrsión Lineal Simple

from sklearn import linear_model

regresion = linear_model.LinearRegression()

# Convertimos las horas a un array de arrays, por tener una sola columna
# array([[40.90547645], [41.57648349], etc])
horas = datos["horas"].values.reshape((-1,1))
modelo = regresion.fit(horas, datos["ingreso"])

print("Intersección (b)", modelo.intercept_)
print("Pendiente (a)", modelo.coef_)

```

```python
# Probamos a hacer una predicción paara esas horas
entrada = [[39], [40], [43], [43.5]]
modelo.predict(entrada)
print(modelo.predict(entrada))
```

```python
# Dibujamos los puntos de la predicción
plt.scatter(entrada, modelo.predict(entrada), color="red")
# Dibujamos la línea de los nuevos salarios y su predicción
plt.plot(entrada, modelo.predict(entrada), color="black")

plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="pink", alpha=0.55)
plt.show()
```

```python
# Otro ejemplo de regresión lineal simple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Genero datos
X = 2*np.random.rand(100, 1)
y = 3 * X + np.random.rand(100, 1)
#Lo convertimos a dataframe
datos = np.concatenate((X, y), axis=1)
df = pd.DataFrame(datos, columns=['X','y'])
df
```

```python
# los dibujamos 
plt.scatter(df['X'], df['y'])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
# xmin, xmax, ymin, ymax: para establecer límites del eje el el gráfico
plt.axis([0, 2, -3, 10])
plt.show()
```

```python
# Creamos el modelo 
from sklearn import linear_model
lin_reg = linear_model.LinearRegression()
# Covertimos a array y tamaño array de arrays
xx = df["X"].values.reshape((-1,1))
lin_reg.fit(xx, df['y'])
print("Intercepto: ", lin_reg.intercept_, ". Coeficiente: ", lin_reg.coef_)
```

```python
# Hacemos una predicción con estos datos, 4 filas una columna
X_new = np.array([[0], [2], [1], [0.5]])
y_predict = lin_reg.predict(X_new)
print(y_predict)
```
